```python
from sklearn import datasets
data=datasets.load_wine()

print(data.keys())
print(data['data'].shape)
print(data['DESCR'])
# print(data['data'])
# print(data['target'])
# print(data['target_names'])
# print(data['feature_names'])
```

## We will be using proline to classify the wine is class0 or above,
### We used the proline as it is the most important feature in the dataset

```python
wine=data['data']
target=data['target']
target_names=data['target_names']

print(wine.shape)
print(target.shape)
print(target_names)
x=wine[:,12:]
y=(target==0).astype(int)





```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression()
clf.fit(x,y)
```

```python
predictions=clf.predict(x)
example_pred=clf.predict(([[800]]))
print(example_pred)

print("The prediction")
for i in range(predictions.shape[0]):
    if predictions[i]==0:
        print("Class 0")
    else:
        print("Above Class 0")
```

```python
score=clf.score(x,y)
print("Accuracy: ",score)
print(x)
```

```python
import numpy as np

l=np.min(x,axis=0)
u=np.max(x,axis=0)

print(l)
print(u)
```

```python
X_new=np.linspace(l,u,1000).reshape(-1,1)
X_prob=clf.predict_proba(X_new)
print(X_prob)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X_new,X_prob[:,1],"r-",label="Class 0")
plt.xlabel("Proline")
plt.ylabel("Probability")
plt.show()
```
