```python
def get_first_name(name):
    return name.split(',')[1]

def get_last_name(name):
    return name.split(',')[0]
df
```

```python

df['First Name'] = df['Name'].apply(get_first_name)
df['Last Name'] = df['Name'].apply(get_last_name)

columns = list(df.columns)


columns.remove('First Name')
columns.remove('Last Name')
columns.remove('Name')

columns.insert(3, 'First Name')
columns.insert(4, 'Last Name')

df = df[columns]

df.to_csv('titanic_2.csv',index=False)
```

```python
df=pd.read_csv('titanic.csv')

import seaborn as sns
import matplotlib.pyplot as plt

print(df.columns)

sns.histplot(df['Age'], bins=10)
plt.show()

sns.barplot(x='Sex', y='Survived', data=df)
plt.show()

numeric_df = df.select_dtypes(include='number')

df_corr = numeric_df.drop(columns='PassengerId').corr()

sns.heatmap(df_corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
