# Why we need sigmoid function?

- In this kernel, I will explain why we need sigmoid function
- Usually in linear regression we use linear function with the form of $y = mx + b$ and loss function is mean squared error
- But in logistic regression, we use sigmoid function with the form of $y = \frac{1}{1 + e^{-x}}$ and loss function is cross entropy
- Why we need sigmoid function and cross entropy loss function in logistic regression?

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


np.random.seed(0)
X = np.random.rand(100, 1) * 10
y = np.sin(X).ravel() + np.random.randn(100) * 0.5  

model = LinearRegression()
model.fit(X, y)

X_test = np.linspace(0, 10, 100).reshape(-1, 1)
y_pred = model.predict(X_test)


plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data with noise')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Linear Regression')
plt.title('Linear Regression on Wiggly Data')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

# Generate synthetic data
X = np.random.rand(100, 1) * 10
y = (X > 5).astype(int).ravel()  


logistic_model = LogisticRegression()
logistic_model.fit(X, y)

# Predictions
y_pred_logistic = logistic_model.predict_proba(X_test)[:, 1]

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Binary Data')
plt.plot(X_test, y_pred_logistic, color='green', linewidth=2, label='Logistic Regression (Sigmoid)')
plt.title('Logistic Regression (Sigmoid) on Binary Data')
plt.xlabel('X')
plt.ylabel('Probability of class 1')
plt.legend()
plt.grid(True)
plt.show()
```

### Key Reasons for Using Cross-Entropy

#### Probabilistic Interpretation:
- Logistic regression outputs probabilities for each class. Cross-entropy loss measures the difference between the predicted probability distribution and the actual distribution of the target class.
- It penalizes incorrect classifications more heavily, especially when the predicted probability is far from the true label.

#### Smooth and Convex:
- Cross-entropy loss is a smooth and convex function, which makes it easier to optimize using gradient descent. Convexity ensures that there is a single global minimum, avoiding issues with local minima.

#### Gradient Properties:
- The gradient of the cross-entropy loss with respect to the weights in logistic regression is directly related to the difference between the predicted probabilities and the actual labels. This property simplifies the gradient descent update rules and leads to efficient training.

### Cross-Entropy Loss Formula
For binary classification, the cross-entropy loss \( L \) is defined as:

L = - (1/m) * Σ [ y(i) * log(p(i)) + (1 - y(i)) * log(1 - p(i)) ]

where:
- m is the number of training examples.
- y(i) is the actual label for the i-th training example (0 or 1).
- p(i) is the predicted probability of the i-th training example being in class 1.
